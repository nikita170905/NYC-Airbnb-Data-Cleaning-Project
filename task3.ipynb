```python
df['name'] = df['name'].fillna('Unknown')
df['host_name'] = df['host_name'].fillna('Unknown')
```

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
```

```python
df['last_review'] = pd.to_datetime(df['last_review'])
df['last_review'] = df['last_review'].fillna(pd.Timestamp('1900-01-01'))
```

```python
df.duplicated().sum()
```

```python
df['room_type'] = df['room_type'].str.strip()
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lb = Q1 - 1.5 * IQR
ub = Q3 + 1.5 * IQR
```

```python
df = df[df['price'] > 0]
df_cleaned = df[(df['price'] >= lb) & (df['price'] <= ub)]
print(f"Shape after outlier removal: {df_cleaned.shape}")
```

```python
print("Cleaned Data Info")
print(df_cleaned.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Before Cleaning (Original Data)
sns.boxplot(ax=axes[0], x=df['price'])
axes[0].set_title('Price Distribution (Before Cleaning)')
axes[0].set_xlabel('Price')

# After Cleaning (Cleaned Data)
sns.boxplot(ax=axes[1], x=df_cleaned['price'])
axes[1].set_title('Price Distribution (After Cleaning)')
axes[1].set_xlabel('Price')

plt.show()
```
